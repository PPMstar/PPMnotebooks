```python
fit_runs = ['D5', 'D6', 'D9', 'D10']
fx = np.log(np.array([run_data[run][1] for run in fit_runs])/1e11)
fy = np.log(np.array([run_data[run][2] for run in fit_runs]))
fc = np.polyfit(fx, fy, 1)
L_fit = np.array((1e10, 1e14))
Mdot_fit = np.exp(fc[0]*np.log(L_fit/1e11) + fc[1])

Mdot0_str = '{:9e}'.format(np.exp(fc[1]))
tmp = Mdot0_str.split('e')
mantissa = float(tmp[0])
exponent = int(tmp[1])
fit_label = r'${:.2f} \times 10^{{{:d}}}$ (L / $10^{{11}}$)$^{{{:.2f}}}$'.\
            format(mantissa, exponent, fc[0])

cb = utils.colourblind
ifig = 1; plt.close(ifig); plt.figure(ifig)
plt.plot(np.log10(L_fit), np.log10(Mdot_fit), ls='-', lw=0.5, color=cb(4), \
         label=fit_label)

i=0
for run in run_data:
    if run.startswith('D') and run_data[run][0] == 768:
        lbl = r'D series, 768$^3$' if i ==0 else ''
        plt.plot(np.log10(run_data[run][1]), np.log10(run_data[run][2]), ls='none', \
                 color=cb(3), marker='o', label=lbl)
        i += 1

i=0
for run in run_data:
    if run.startswith('D') and run_data[run][0] == 1536:
        lbl = r'D series, 1536$^3$' if i ==0 else ''
        plt.plot(np.log10(run_data[run][1]), np.log10(run_data[run][2]), ls='none', \
                 color=cb(4), marker='v', label=lbl)
        i += 1

plt.plot(np.log10(run_data['I12'][1]), np.log10(run_data['I12'][2]), ls='none', \
         markeredgecolor=cb(1), markerfacecolor='none', markeredgewidth=1, \
         marker='H', label='I12')

i=0
for run in run_data:
    if run.startswith('I') and run_data[run][0] == 768 and run_data[run][4] == 1:
        lbl = r'I series, 768$^3$, Net 1' if i ==0 else ''
        plt.plot(np.log10(run_data[run][1]), np.log10(run_data[run][2]), ls='none', \
                 color=cb(1), marker='s', label=lbl)
        i += 1
        
i=0
for run in run_data:
    if run.startswith('I') and run_data[run][0] == 768 and run_data[run][4] == 2:
        lbl = r'I series, 768$^3$, Net 2' if i ==0 else ''
        plt.plot(np.log10(run_data[run][1]), np.log10(run_data[run][2]), ls='none', \
                 color=cb(8), marker='^', label=lbl)
        i += 1

i=0
for run in run_data:
    if run.startswith('I') and run_data[run][0] == 1152 and run_data[run][4] == 1:
        lbl = r'I series, 1152$^3$, Net 1' if i ==0 else ''
        plt.plot(np.log10(run_data[run][1]), np.log10(run_data[run][2]), ls='none', \
                 color=cb(4), marker='*', label=lbl)
        i += 1
        
plt.xlabel(r'log$_{10}$(L / L$_\odot\!$)')
plt.ylabel(r'log$_{10} ( \dot{\mathrm{M}}_\mathrm{e}$ / M$_\odot$ s$^{-1}$])')
plt.xlim((10., 13.49))
plt.ylim((-7.5, -3.5))

handles, labels = plt.gca().get_legend_handles_labels()
legend2 = plt.legend(handles[3:], labels[3:], loc=4)
plt.legend(handles[0:3], labels[0:3], loc=2)
plt.gca().add_artist(legend2)

plt.tight_layout()
plt.savefig('entrainment_rate_vs_luminosity.pdf')
```
